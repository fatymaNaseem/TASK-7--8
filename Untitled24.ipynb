```python
from google.colab import files
import pandas as pd
import matplotlib.pyplot as plt

# Upload the file
uploaded = files.upload()

# Load the dataset
file_path = list(uploaded.keys())[0]
data = pd.ExcelFile(file_path)

# Load the "PK" sheet for national data
pk_data = data.parse('PK')

# Check for null values
pk_data.fillna(0, inplace=True)

# Filter data up to September 1, 2020
filtered_data = pk_data[pk_data['Date'] <= '2020-09-01']

# Total tests
total_tests = filtered_data['Daily Tests'].sum()
print(f"Total Tests: {total_tests}")

# Trend Line, Scatter Plot, Bar Graph for Daily Tests
plt.figure(figsize=(15, 10))

# Trend Line
plt.subplot(3, 1, 1)
plt.plot(filtered_data['Date'], filtered_data['Daily Tests'], label='Daily Tests', color='blue')
plt.title('Trend Line: Daily Tests')
plt.xlabel('Date')
plt.ylabel('Tests')
plt.legend()

# Scatter Plot
plt.subplot(3, 1, 2)
average_tests = filtered_data['Daily Tests'].mean()
above_avg = filtered_data['Daily Tests'] > average_tests
below_avg = ~above_avg
plt.scatter(filtered_data['Date'][above_avg], filtered_data['Daily Tests'][above_avg], color='red', label='Above Average')
plt.scatter(filtered_data['Date'][below_avg], filtered_data['Daily Tests'][below_avg], color='green', label='Below Average')
plt.title('Scatter Plot: Daily Tests (Above/Below Average)')
plt.xlabel('Date')
plt.ylabel('Tests')
plt.legend()

# Bar Graph
plt.subplot(3, 1, 3)
plt.bar(filtered_data['Date'], filtered_data['Daily Tests'], color='cyan')
plt.title('Bar Graph: Daily Tests')
plt.xlabel('Date')
plt.ylabel('Tests')

plt.tight_layout()
plt.show()

# Convert data to weekly and monthly statistics
weekly_data = filtered_data.resample('W-Mon', on='Date').sum()
monthly_data = filtered_data.resample('M', on='Date').sum()

# Save weekly and monthly statistics
weekly_data.to_csv('weekly_stats.csv')
monthly_data.to_csv('monthly_stats.csv')

# Find max and min for weekly and monthly data
weekly_max_tests = weekly_data['Daily Tests'].idxmax(), weekly_data['Daily Tests'].max()
weekly_min_tests = weekly_data['Daily Tests'].idxmin(), weekly_data['Daily Tests'].min()
monthly_max_tests = monthly_data['Daily Tests'].idxmax(), monthly_data['Daily Tests'].max()
monthly_min_tests = monthly_data['Daily Tests'].idxmin(), monthly_data['Daily Tests'].min()

print(f"Weekly Max Tests: {weekly_max_tests}")
print(f"Weekly Min Tests: {weekly_min_tests}")
print(f"Monthly Max Tests: {monthly_max_tests}")
print(f"Monthly Min Tests: {monthly_min_tests}")

# Per Thousand and Per Million Calculations
population = 220000000  # Assuming Pakistan's population is 220 million
filtered_data['Tests per Thousand'] = (filtered_data['Daily Tests'] / population) * 1000
filtered_data['Cases per Thousand'] = (filtered_data['Daily Cases'] / population) * 1000
filtered_data['Deaths per Thousand'] = (filtered_data['Daily Deaths'] / population) * 1000

# Train ML Models for Predictions
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor

# Prepare data
X = filtered_data.index.values.reshape(-1, 1)
y_cases = filtered_data['Daily Cases']
y_deaths = filtered_data['Daily Deaths']

# Train/Test Split
X_train, X_test, y_train_cases, y_test_cases = train_test_split(X, y_cases, test_size=0.2, random_state=42)
X_train, X_test, y_train_deaths, y_test_deaths = train_test_split(X, y_deaths, test_size=0.2, random_state=42)

# Linear Regression
lr = LinearRegression()
lr.fit(X_train, y_train_cases)
predicted_cases = lr.predict(X_test)

# Decision Tree
dt = DecisionTreeRegressor()
dt.fit(X_train, y_train_cases)
predicted_cases_dt = dt.predict(X_test)

print("Linear Regression and Decision Tree Models Trained Successfully")
```
